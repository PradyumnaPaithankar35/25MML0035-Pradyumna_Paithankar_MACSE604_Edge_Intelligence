# **Import** **Libraries**

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

# **Load Data**
# 1. Data Jar
The data used here is MNIST Digit data

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

# **Shape of Data**

```python
print("X train shape: ", x_train.shape)
print("Y train shape: ", y_train.shape)
print("X test shape: ",x_test.shape)
print("Y test shape: ",y_test.shape)
```

```python
import matplotlib.pyplot as plt
```

# **Inspect Data**

```python
plt.figure(figsize=(10,10))
for i in range(9):
    plt.subplot(3,3, i+1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(f"Label: {y_train[i]}")
    plt.axis("off")
plt.show()
```

# **Normalize Data**

```python
x_train = x_train.astype("float32")/255.0
x_test = x_test.astype("float32")/255.0
```

# **Train Model**
#2. Task Jar
The task is to classify the hand written digits.
#3. Model Jar
The model used is Artificial Neural Network (ANN)
#4. Loss Jar
The loss used is sparse_categorical_crossentropy

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

model = Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer = 'sgd', loss = 'sparse_categorical_crossentropy',metrics = ["accuracy"])
```

#5. Learning Jar

```python
model.fit(x_train, y_train, epochs = 20, batch_size = 64)
```

# **Model Summary**

```python
model.summary()
```

# **Test Accuracy**
#6. Accuracy Jar

```python
loss, acc = model.evaluate(x_test, y_test)
print("Accuraccy: ",acc*100)
print("Loss: ",loss)
```

# **Pickle the Model**

```python
import pickle

with open("mnist.pkl", "wb") as file:
    pickle.dump(model, file)
```
